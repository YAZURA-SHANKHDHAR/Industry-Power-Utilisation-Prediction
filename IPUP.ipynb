```python
def preprocessing(dt):
  if'Date' in dt.columns:
    dt=dt.drop(['Unnamed: 0','Date'],axis=1)
  else:
    dt=dt.drop(['Unnamed: 0'],axis=1)
  dt.head()
  dt['month'] = dt['date'].dt.month
  dt['day'] = dt['date'].dt.day
  dt['day_of_year'] = dt['date'].dt.day_of_year

  for val in dt['Duty']:
    if val == 'Low':
      dt['Duty'] = dt['Duty'].replace(val, 0)
    elif val == 'Average':
      dt['Duty'] = dt['Duty'].replace(val, 1)
    elif val == 'Major':
      dt['Duty'] = dt['Duty'].replace(val, 2)
  return dt
```

```python
test = test.rename(columns={'TimeStamp':'date'})
test.date=pd.to_datetime(test.date, infer_datetime_format=True)
train.date=pd.to_datetime(train.date, format="%d-%m-%Y %H:%M")
```

```python
train = preprocessing(train)
test = preprocessing(test)
```

```python
y = train.pop('Energy_Consumption')
```

```python
x = train
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
```

```python
def rmse(ytest, ypred):
  return np.sqrt(mean_squared_error(ytest, ypred))
```

```python
rnd = RandomForestRegressor()
rnd.fit(xtrain, ytrain)
rnd_rmse = rmse(ytest, rnd.predict(xtest))
rnd_rmse
```

```python
xgb = xgb.XGBRegressor()
xgb.fit(xtrain, ytrain)
xgb_rmse = rmse(ytest, xgb.predict(xtest))
xgb_rmse
```

```python
dtree = DecisionTreeRegressor()
dtree.fit(xtrain, ytrain)
dtree_rmse = rmse(ytest, dtree.predict(xtest))
dtree_rmse
```

```python
grdr = GradientBoostingRegressor()
grdr.fit(xtrain, ytrain)
grdr_rmse = rmse(ytest, grdr.predict(xtest))
grdr_rmse
```
